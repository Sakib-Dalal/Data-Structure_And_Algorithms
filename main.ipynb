<h3> 1. O(n) </h3>



<p> The graph shows as the number of Elements increases, then the number of Operations also increases. </p>

```python
import time

arr = ['nemo', 'dory', 'bruce', 'cat', 'dog', 'tiger']

def func(my_arr):
    t0 = time.time()
    for i in my_arr:
        if i == 'tiger':
            print("Found Tiger")
    t1 = time.time()
    print("Call to find Tiger took: ", t1-t0, " mill seconds")

func(arr) # O(n) ==> Linear Time
```

<p> We can see in the given example a pattern of O(n). We can say that it takes linear time to find tiger. </p>
<p> Consider an example where a n'th number of boxes are used to compressed. The function will compress boxes one by one which will take 'T' time to complete all box compression. </p>

<h3> 2. O(1) </h3>



<p> Now consider an example where you want to operate only one expresion as given in the code below. We call O(1) as Constant Time. </p>

```python
import time

arr = ['nemo', 'dory', 'bruce', 'cat', 'dog', 'tiger']

def func(my_arr):
    t0 = time.time()

    print(my_arr[0]) # O(1) ==> operate only one expression, we call this constant time
    
    t1 = time.time()
    print("time took: ", t1-t0, " mill seconds")

func(arr)
```

<p> Now what if we want to operate two element from the given array. </p>

```python
import time

arr = ['nemo', 'dory', 'bruce', 'cat', 'dog', 'tiger']

def func(my_arr):
    t0 = time.time()

    print(my_arr[0]) # O(1) ==> First operation
    print(my_arr[1]) # O(n) ==> Second operation
    
    t1 = time.time()
    print("time took: ", t1-t0, " mill seconds")

func(arr) # O(2)
```

<p> We can say that this function is running operation of 2 elements every time. We can increase the number of operation by increasing elements but every time it'a a flat line on grap. </p>

<h5> Exercise: Calculate the Big O notation. </h5>

```python
import time

my_arr = [1, 2, 3, 4, 5]

def func(input):
    t1 = time.time() # O(1) --> 1.
    a = 10 # O(1) --> 2.
    a = 50 + 3 # O(1)  --> 3.

    for i in range(len(input)):  # O(n)  --> 4.
        var = True # O(1) * n => O(n)  --> 5.
        a += 1 # O(1) * n => O(n)  --> 6.
    t2 = time.time() # O(n)  --> 7.
    print(f"Time took: {t2 - t1} msec.") # O(n)  --> 8.
    return a # O(1)  --> 9.

func(my_arr)
```

<p> So if we calculate the efficiency of each Big O it turn out to be: 3 + 3n + 2 = 5 + 3n --> <b>BIG O(5 + 3n)</b> </p>

<h3> Simplifying Big O</h3>

```python
def funchallenge(input):
    temp = 10 #O(1)
    temp = temp +50 #O(1)
    for i in range(len(input)): #O(n)
        var = True #n*O(1)
        temp += 1 #n*O(1)
    return temp #O(1)

funchallenge(my_arr)
```

<p>Total running time of the funchallenge function =
O(1 + 1 + n + n*1 + n*1 + n*1 + 1) = O(3n +3) = O(3(n+1))
<u>Any constant in the Big-O representation can be replaced by 1, as it doesn't really matter what constant it is.
Therefore, O(3(n+1)) becomes O(n+1)
Similarly, any constant number added or subtracted to n or multiplied or divided by n can also be safely written as just n
This is because the constant that operates upon n, doesn't depend on n, i.e., the input size
Therefore, the funchallenge function can be said to be of O(n) or Linear Time Complexity.</u></p>

<h4>Worst Case</h4>
<p>Consider an example of finding nemo where we constantrly run function to find nemo in given array, but after finding nemo we till run the function constantly untill the item's finish. This can be time consuming and extra operations but we can break the loop of the function after it satisfies an operation as follown in the example's.</p>

```python
# Non-efficient code 

import time

arr = ['dory', 'bruce', 'nemo', 'cat', 'dog', 'tiger', 'lion']

def func(my_arr):
    t0 = time.time()
    for i in my_arr:
        print("running")
        if i == 'nemo':
            print("Found Nemo")
    t1 = time.time()
    print("Time took: ", t1-t0, " mill seconds")

func(arr)
```

```python
# efficient code 

import time

arr = ['dory', 'bruce', 'nemo', 'cat', 'dog', 'tiger', 'lion']

def func(my_arr):
    t0 = time.time()
    for i in my_arr:
        print("running")
        if i == 'nemo':
            print("Found Nemo")
            break # --> using break to exit the loop
    t1 = time.time()
    print("Time took: ", t1-t0, " mill seconds")

func(arr)
```

<h3>3. O(n^2)</h3>
<p>Q. What is the Big O notation of the given code?</p>

```python
arr = [1, 2, 3]

def logAllPairs(arr):
    for i in arr: # --> O(n)
        for j in arr: # --> O(n * n) -> O(n ^ 2)
            print(i, j)

logAllPairs(arr) # O(n^2) --> O(n * n) --> O(a * b)
```

```python
# Example 2

array = ['a','b','c','d','e']

def log_all_pairs(array):

    #There are nested for loops in this function but there is only one variable array. So we don't need two variables for the Big-O

    for i in range(len(array)): #n*O(n)
        for j in range(len(array)): #n*O(n)
            print(array[i], array[j]) #n*n*O(1)

log_all_pairs(array)

#Total time complexity of the log_all_pairs function =
#O(n*n + n*n + n*n*1) = O(3n^2)
#The constants can be safely ignored.
#Therefore, O(3n^2) = O(n^2)
```

<p>We can notice that the code consiste a nested for loop. We can see two n's nested in for loop, hence <b>Big O will be O(n * n) or we can write it like O(n^2). <u>This O(^2) we called it as Quadratic Time.</u></b> That means every time number of elements increased the number of operation also increases.</p>

<h4>Drop Non Dominants</h4>
<p>Consider an example given bellow:</p>

```python
new_array = [1,2,3]
def print_numbers_then_pairs(array):

    #There are a total of three loops here but only one variable. So we need only variable for our Big-O notation

    print("The numbers are : ") #O(1)
    for i in range(len(array)): #O(n)
        print(array[i]) #n*O(1)

    print("The pairs are :") #O(1)
    for i in range(len(array)): #n*O(n)
        for j in range(len(array)): #n*O(n)
            print(array[i], array[j]) #n*n*O(1)

print_numbers_then_pairs(new_array) # --> O(n + n^2) Now we always drop the non dominant term so we drop n -> O(n + n ^2) = O(n^2)
```

<p>Total time complexity of the print_numbers_then_pairs function =
O(1 + n + n*1 + 1 + n*n + n*n + n*n*1) = O(3n^2 + 2n + 2)</p>
<p>Now, Big-O presents scalability of the cod, i.e., how the code will behave as the inputs grow larger and larger
Therefore if the expression contains terms of different degrees and the size of inputs is huge, the terms of the smaller degrees become negligible in comparison to those of the higher degrees</p>
<p>Therefore, we can ignore the terms of the smaller degrees and only keep the highest degree term
O(3n^2 + 2n + 2) = O(3n2)</p>
<p>The constants can be safely ignored.
Therefore, O(3n^2) = O(n^2)</p>


<h3>Space Complexity</h3>
<p>Consider a function which comprise boxes. The function do it's job in a Time Complexity but how many boxes can it take and store to do its task is based on it's memory and Space complexity.<p>

<h5>What causes Space Complexity?</h5>
<ul>
    <li>Variables</li>
    <li>Data Structure</li>
    <li>Function Call</li>
    <li>Allocations</li>
</ul>

```python
# Variable Space Complexity

def func(n):
    for i in n:  # here i is alocated a memory which is Variable Space Complexity
        print("booo!")

func([1, 2, 3, 4, 5]) # O(1)
```

```python
# Variable and Data Structure Space Complexity

def func(n):
    my_lst = [] # here we alocated memory which is Variable Space Complexity
    for i in n: # i is also taking Space
        my_lst.append("hi") # appending hi to list which is Data Structure Space Complexity
    print(my_lst)

func([1, 2, 3, 4, 5])  # O(n)
```

<h2>Efficient and Non-efficient code example</h2>

```python
'''Given 2 arrays, create a function that let's a user know (true/false)
whether these two arrays contain any common items'''
'''For example
arr1 = ['a', 'b', 'c', 'x']
arr2 = ['z', 'y', 'i']
should return false

arr1 = ['a', 'b', 'c', 'x']
arr2 = ['z', 'y', 'x']
should return true'''

arr1 = ['a', 'b', 'c', 'x']
arr2 = ['z', 'y', 'x']

# non efficient code 
def func(arr1, arr2):
    for i in range(0, len(arr1)):
        for j in range(0, len(arr2)):
            if arr1[i] == arr2[j]:
                return True                
            
func(arr1, arr2) # O(n^2) Notation
```

```python
# efficient code

arr1 = ['a', 'b', 'c', 'x']
arr2 = ['z', 'y', 'x']


def func(arr1, arr2):
    for j in arr2:
        if j in arr1:
            return True
    else:
        return False
        
func(arr1, arr2) # O(n)
```
